```python
import matplotlib.dates as mdates
# convert the datetime index to a numeric index

# assume your time series is stored in a variable called 'ts'
result = seasonal_decompose(data['Oslo'], model='additive', period=24)

# extract the trend, seasonal and residual components
trend = result.trend
seasonal = result.seasonal
residual = result.resid


x = mdates.date2num(data.index.to_pydatetime())

# create a new figure with a larger size
fig, ax = plt.subplots(figsize=(14, 2))

# plot the seasonal component with a dotted line
ax.plot(x, seasonal, linestyle='--', label='Seasonal')

# set the x-axis to display dates in a readable format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# set the x-axis tick frequency to daily
ax.xaxis.set_major_locator(mdates.DayLocator())

# set the x-axis limits to display one week of data
ax.set_xlim([mdates.date2num(data.index[0]), mdates.date2num(data.index[96])])
ax.set_ylabel('Seasonal')

# display the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, trend)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.set_xlim([mdates.date2num(data.index[0]), mdates.date2num(data.index[-1])])
ax.set_ylabel('Trend')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, residual)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.set_xlim([mdates.date2num(data.index[0]), mdates.date2num(data.index[-1])])
ax.set_ylabel('Residual')
plt.show()
```
