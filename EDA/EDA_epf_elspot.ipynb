```python
from sklearn.decomposition import PCA
df1.drop(['datetime', 'target', 'da_production'], 1, inplace=True)
```

```python
# Calculate the covariance matrix

cov_mat = df1.corr()

res=[]
for i in range(1, 6):
    pca = PCA(n_components=i)
    pca.fit(cov_mat)
    res.append(pca.explained_variance_ratio_.sum())

# Plot the elbow point in the explained variance ratios
plt.plot(range(1, 6), res, marker='o', color='g')
plt.xlabel('Number of components')
plt.ylabel('Explained variance ratio')
plt.title('Elbow point in explained variance')
plt.show()
```

```python
# Fit PCA model
pca = PCA(n_components=2)
principal_components = pca.fit_transform(cov_mat)

# Plot the principal components
plt.figure(figsize=(10,8))
plt.scatter(principal_components[:, 0], principal_components[:, 1], s=60, c=[i for i in range(12)])

# Annotate each point with its variable name
variables = df1.columns
for i in range(min(len(variables), 12)):
    plt.annotate(variables[i], xy=(principal_components[i, 0], principal_components[i, 1]), fontsize=13)

# Set axis labels
plt.xlabel('PC1 ({:.2f}%)'.format(100*pca.explained_variance_ratio_[0]))
plt.ylabel('PC2 ({:.2f}%)'.format(100*pca.explained_variance_ratio_[1]))
plt.show()
```

```python
n_pcs= pca.n_components_ # get number of component
# get the index of the most important feature on EACH component
most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
initial_feature_names = df1.columns
# get the most important feature names
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]
```
