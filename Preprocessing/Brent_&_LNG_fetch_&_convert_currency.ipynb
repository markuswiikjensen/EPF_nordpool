```python
import yfinance as yf
data = yf.download("BZ=F NG=F", start="2019-12-31", end="2023-4-1", interval="1d")
```

```python
data= data['Adj Close']
```

```python
data['Date']=data.index.copy()
```

```python
from tqdm import tqdm
from tqdm.notebook import trange, tqdm

tqdm.pandas(desc='Converting currency to NOK')

from forex_python.converter import CurrencyRates
import pandas as pd

# Create a CurrencyRates object to get exchange rates
c = CurrencyRates()

# Read the historical prices from a CSV file
prices = data[['BZ=F','NG=F','Date']]

# Define the source and target currencies
source_currency = "USD"
target_currency = "NOK"

# Loop over the prices and convert them to the target currency
for index, row in tqdm(prices.iterrows()):
    date = row["Date"]
    price1 = row["BZ=F"]
    price2= row["NG=F"]
    try:
        exchange_rate = c.get_rate(source_currency, target_currency, date)
        converted_price1 = price1 * exchange_rate
        converted_price2 = price2 * exchange_rate
        prices.at[index, "BZ=F"] = converted_price1
        prices.at[index, "NG=F"] = converted_price2
    except:
        pass



# Save the converted prices to a new CSV file
prices.to_csv("prices_brent_nok.csv", index=False)
```
